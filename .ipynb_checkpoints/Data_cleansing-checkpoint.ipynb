### Addressing null values 

I decided to forward fill the null values in the Country/registry and Anomaly columns because when the data was read in from the xls format, these values were coded as nan when originally they were merged cells.

```python
df['Country / registry'].replace('nan', np.nan, inplace=True)
```

```python
df['Country / registry'].fillna(method='ffill', inplace=True)
```

```python
df.replace('Pleven - Bulgaria ¹', 'Pleven - Bulgaria', inplace = True)
#removing numerical figure in Pleven, Bulgaria name.
```

### Removing foreign characters

* Removed confidence interval values that were in the brakcets, using RegEx, and kept the avarage values that were already stated.

```python
df.columns = df.columns.map(lambda x: re.sub(r"\(.*\)", "", x))
```

```python
df.head()
```

```python
df['total_prevalence_'] = df['total_prevalence_'].apply(lambda x: re.sub(r"\(.*\)", "", x))
```

```python
df['lb_prevalence_'] = df['lb_prevalence_'].apply(lambda x: re.sub(r"\(.*\)", "", x))
```

```python
df['fd_prevalence_'] = df['fd_prevalence_'].apply(lambda x: re.sub(r"\(.*\)", "", x))
```

```python
df['topfa_prevalence_'] = df['topfa_prevalence_'].apply(lambda x: re.sub(r"\(.*\)", "", x))
```

```python
df['lb_proportion_'] = df['lb_proportion_'].apply(lambda x: re.sub(r"\(.*\)", "", x))
```

```python
df['still_birth_proportion_'] = df['still_birth_proportion_'].apply(lambda x: re.sub(r"\(.*\)", "", x))
```

```python
df['topfa_proportion_'] = df['topfa_proportion_'].apply(lambda x: re.sub(r"\(.*\)", "", x))
```

```python
df.head()
```

```python
df[['lb_proportion_', 'still_birth_proportion_', 'topfa_proportion_']] = df[['lb_proportion_', 'still_birth_proportion_', 'topfa_proportion_']].applymap(lambda x: x.replace('%', ''))
```

```python
df[['lb_proportion_', 'still_birth_proportion_', 'topfa_proportion_']] = df[['lb_proportion_', 'still_birth_proportion_', 'topfa_proportion_']].replace('-', 0)
```
